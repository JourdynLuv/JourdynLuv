```python
import pandas as pd

url = 'https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv'
df = pd.read_csv(url)

df.head()

top_attackers = df.sort_values(by='Attack', ascending=False).head(5)

top_attackers[['Name', 'Attack', 'Type 1', 'Legendary']]
```
