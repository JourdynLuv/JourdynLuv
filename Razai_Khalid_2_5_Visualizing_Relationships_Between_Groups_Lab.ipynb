# Lab Instructions

You are part of a data team for a financial institution that uses an AI model to determine who will and will not be granted a loan.  Recently, there have been complaints from customers that the model unfairly discriminates based on the applicant's age and sex. A court has ordered that the financial institution must provide information about how the model works.  

Data from recent loan applications is given in `loan_data.csv`.  The data dictionary is below:

| Feature              | Description | Data Type |
| :---------------- | :------: | ----: |
|person_age|Age of the person|Float
|person_gender|Gender of the person|Categorical
|person_education|Highest education level|Categorical
|person_income|Annual income|Float
|person_emp_exp|Years of employment experience|Integer
|person_home_ownership|Home ownership status (e.g., rent, own, mortgage)|Categorical
|loan_amnt|Loan amount requested|Float
|loan_intent|Purpose of the loan|Categorical
|loan_int_rate|Loan interest rate|Float
|loan_percent_income|Loan amount as a percentage of annual income|Float
|cb_person_cred_hist_length|Length of credit history in years|Float
|credit_score|Credit score of the person|Integer
|previous_loan_defaults_on_file|Indicator of previous loan defaults|Categorical
|loan_status (target variable)|Loan approval status: 1 = approved; 0 = rejected|Integer

Visualize the relationship between age, sex and 5 other features with `loan_status`.  Which features seem to drive the model's decision to grant a loan?  Which seem unrelated?  Should the financial institution be concerned about potential legal trouble?  Make sure to justify your answers with specific references to your visualizations.

**You must use quantitative visualizations (boxplots or histograms) when comparing quantitative features.  DO NOT categorize quantitative features.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content//sample_data/loan_data.csv')

df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='person_age', data=df)
plt.title('Distribution of Applicant Age by Loan Status')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='person_gender', y='loan_status', data=df)
plt.title('Loan Approval Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='person_income', data=df)
plt.title('Distribution of Income by Loan Status')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='credit_score', data=df)
plt.title('Distribution of Credit Score by Loan Status')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='person_emp_exp', data=df)
plt.title('Employment Experience (Years) by Loan Status')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='person_home_ownership', y='loan_status', data=df)
plt.title('Loan Approval Rate by Home Ownership')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='previous_loan_defaults_on_file', y='loan_status', data=df)
plt.title('Loan Approval Rate by Previous Defaults')
plt.show()
```

**Which features seem to drive the model's decision to grant a loan? **

One of the biggest drivers if you look at the barplot is applicants with no had a higher approval rate at 12%, while people that had yes as on the report had 0% approval rate.

Another feature was Home ownership, people who own their home had almost 25% approval rate which was vastly higher than those who rent or had a mortgage.

Finally income had a median of around $75k had a higher approval rate than the median of around $60k which was more rejected.

**Which seem unrelated? **

Employment Experience seemed to be most unrelated as if you look at the bar plot, the boxes are around the same height and the median line for both approvaed and rejected is almost the same and the model did not read much into it.

**Should the financial institution be concerned about potential legal trouble?**

Yes they should be very converned as the gender barplot shows no evidence of sex discrimination, there is clearly evidence of age discrimination. The age box plot shows the median age for rejected applicants is around 55 while the median for approved drops to 39 showing it is clearly favoring young applicants.
