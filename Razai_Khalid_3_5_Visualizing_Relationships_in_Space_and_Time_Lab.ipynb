The data set I am using is the share of population using the internet, the data comes from the international telecom union. The dataset contains the spatial data and the time series data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/share-of-individuals-using-the-internet.csv')

df = df.rename(columns={'Individuals using the Internet (% of population)': 'Internet_Usage'})
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='year', y='it_net_user_zs', alpha=0.3, color='blue')

plt.title("Global Internet Usage Over Time (All Countries)")
plt.ylabel("Percentage of Population Using Internet")
plt.show()
```

The scatterplot shows the globael trend of internet adoption. by plotting every countrys internet usage over time, you can see that before 2000, global usage was very low, and over the last few decades has risen all over the world


```python
countries_to_compare = ['United States', 'Thailand', 'World']
df_compare = df[df['entity'].isin(countries_to_compare)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_compare, x='year', y='it_net_user_zs', hue='entity', linewidth=2)
plt.title("Internet Adoption: United States vs. Thailand vs. World Average")
plt.ylabel("Percentage of Population Using Internet")
plt.show()
```

THe line chart compares the spatial differences in internet adoption between the USA, Thailand and the global world average. It can be seen that the US adopted internet early and steadily while Thailand was delayed but massive spike around 2005-2010


```python
diverse_countries = ['United States', 'Mexico', 'India', 'Nigeria']
df_boxes = df[(df['entity'].isin(diverse_countries)) & (df['year'] >= 2000)]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_boxes, x='entity', y='it_net_user_zs')
plt.title("Spatial Distribution of Internet Usage (2000 - 2020)")
plt.ylabel("Percentage of Population Using Internet")
plt.xlabel("Geographical Space (Country)")
plt.xticks(rotation=45)
plt.show()
```

This boxplot shows the differences in internet adoption across four geographic spaces over 20 year span. While the US maintains a high concentration of usage with a high median while countries like India and Nigeria show much lower overall medians, showing they had a slower adoption rate. While mexico is very spread out and shows rapid growth and a difference in the two decades.
